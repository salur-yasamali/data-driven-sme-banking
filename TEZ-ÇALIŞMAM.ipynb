```python
df = df.drop(columns= ['balance','k_symbol','bank','account']) 
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%y%m%d') 
df.head()
```

```python
df['type'].unique()
```

```python
df =df[df['type']=='PRIJEM']
df.info()
```

```python
rfm_df = df.groupby('account_id').agg({'amount' : 'sum' , 'date' : ['count','max']})
rfm_df.head()
```

```python
rfm_df.columns = ['Monetary', 'Frequency', 'Last_Date']
rfm_df.head(10)
```

```python
analiz_tarihi = df['date'].max() + pd.Timedelta(days=1)
rfm_df['Recency'] = (analiz_tarihi - rfm_df['Last_Date']).dt.days
rfm_df.tail(10)
```

```python
rfm_df = rfm_df.drop(columns='Last_Date') 
rfm_df.head()
```

```python
rfm_df['Recency_Score'] = pd.qcut(rfm_df['Recency'].rank(method='first'), q=5, labels=[5,4,3,2,1])    #rank methodu 5 e tam bölünmediği için ekledik, bölünüyomuş gibi yapsın diye
rfm_df['Frequency_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5])
rfm_df['Monetary_Score'] = pd.qcut(rfm_df['Monetary'].rank(method='first'), q=5, labels=[1,2,3,4,5])
rfm_df.head()
```

```python
rfm_df['Recency_Score']= rfm_df['Recency_Score'].astype(str) 
rfm_df['Frequency_Score']= rfm_df['Frequency_Score'].astype(str)    # skorları yan yana yazdırabilmek için str yaptık, toplamasın yan yana yazsın diye
rfm_df['Monetary_Score']= rfm_df['Monetary_Score'].astype(str)   

rfm_df['RFM_skoru'] = rfm_df['Recency_Score'] + rfm_df['Frequency_Score'] + rfm_df['Monetary_Score'] 
rfm_df.head()
```

```python
rfm_df['RF_Score'] = rfm_df['Recency_Score'] + rfm_df['Frequency_Score'] 
rfm_df.head()    # rfm de kategoriler genelde rf ye göre yapılır, 5x5 lik matrise uysun diye, m si daha sonra yine kullanılır.
```

```python
seg_map = {
    r'[1-2][1-2]': 'Uyuyanlar',
    r'[1-2][3-4]': 'Riskli',
    r'[1-2]5': 'Kaybedilemez',
    r'3[1-2]': 'Uyku_Adayi',
    r'33': 'Dikkat_Gereken',
    r'[3-4][4-5]': 'Sadik_Musteriler',
    r'41': 'Umut_Veren',
    r'51': 'Yeni_Musteriler',
    r'[4-5][2-3]': 'Potansiyel_Sadik',
    r'5[4-5]': 'Sampiyonlar'
}

rfm_df['Segment']= rfm_df['RF_Score'].replace(seg_map, regex=True) 
rfm_df.tail(10)
```

```python
rfm_df.groupby('Segment').agg({'Recency': ['count','mean'] , 'Frequency': 'mean','Monetary' : 'mean'})  
# segmentlerin özetini çıkardık, hangi segment bize ne katıyor.
```

```python
total_customers = rfm_df.shape[0]
total_monetary = rfm_df['Monetary'].sum()
pareto_df = rfm_df.groupby('Segment').agg(Müşteri_Sayısı=('Recency', 'count'),Toplam_Ciro=('Monetary', 'sum')).reset_index()

pareto_df
```

```python
pareto_df['Müşteri_yüzdesi'] = (pareto_df['Müşteri_Sayısı']/ total_customers) *100
pareto_df['Ciro_yüzdesi'] = (pareto_df['Toplam_Ciro']/ total_monetary) * 100 

pareto_df = pareto_df.sort_values(by='Ciro_yüzdesi', ascending=False) 
pareto_df
```

```python
ozet_tablo = rfm_df.groupby('Segment').agg({
    'Recency': ['count', 'mean'], 
    'Frequency': 'mean', 
    'Monetary': 'mean'
})


ozet_tablo.to_excel('RFM_Ozet_Tablosu.xlsx')


pareto_df.to_excel('Pareto_Analizi.xlsx', index=False)
```
